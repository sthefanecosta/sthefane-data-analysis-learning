### Exercício 07: Listas - Acesso por Índice
1. Crie uma lista com pelo menos 4 disciplinas de sua preferência.
2. Imprima a primeira disciplina da lista utilizando o índice correspondente.
3. Imprima a última disciplina da lista utilizando o índice negativo (-1).
4. Imprima a segunda disciplina da lista.

```python
# Criando a lista
disciplinas = ['artes', 'biologia', 'fisíca', 'computação']

# Imprimindo o primeiro item da lista
print(disciplinas[0])

# Imprimindo o último item da lista
print(disciplinas[-1])

# Imprimindo o segundo item da lista
print(disciplinas[1])
```

### Exercício 08: Listas - Modificação
1. Crie uma lista com 3 tarefas diárias (ex: "Estudar Python", "Fazer exercícios", "Dormir cedo").
2. Adicione uma nova tarefa ao final da lista utilizando o método `append()`.
3. Remova um item da lista utilizando o método `remove()`.
4. Altere o valor da primeira tarefa da lista para um novo valor e imprima a lista final.

```python
# Criando a lista
tarefas = ['Estudar Python', 'Fazer exercícios', 'Dormir cedo']

# Adicionando uma nova tarefa
tarefas.append('Tomar banho')

# Removendo um item da lista
tarefas.remove(tarefas[1])

# Alterando o primeiro item da lista
tarefas[0] = 'Acordar'

# Exibindo a lista
print(tarefas)
```

### Exercício 09: Loops - Percorrendo Listas
1. Crie uma lista com preços de produtos: `[50, 150, 30, 200]`.
2. Utilize um loop `for` para percorrer a lista e imprimir cada preço formatado como "R$ X".
3. Calcule o valor total da lista utilizando a função `sum()`.
4. Imprima a quantidade total de itens da lista utilizando a função `len()`.

```python
# Criando a lista de preços
precos = [50, 150, 30, 200]

# Utilizando o loop para percorrer a lista
for x in precos:
    print(f'R$ {x:.2f}')

# Calculando o valor total da lista
valor_total = sum(precos)
print(f'Valor total: R$ {valor_total:.2f}')

# Exibindo a quantidade total de itens
print(len(precos))
```
