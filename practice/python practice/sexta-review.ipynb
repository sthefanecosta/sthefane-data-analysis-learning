**Descrição:** Nesta etapa, consolidamos o aprendizado da semana criando um sistema funcional. O desafio exige a integração de dicionários, listas, entrada de dados (input), processamento lógico (cálculo de média e condicionais) e manipulação de strings para a exibição de um relatório.

### Desafio Integrador: Sistema de Estudante
Crie um programa completo que interage com o usuário para coletar dados, armazena essas informações em um dicionário estruturado e processa um relatório de desempenho escolar.

```python
# Pedindo os dados ao usuário
nome = input('Nome do aluno(a): ')
idade = int(input('Idade do aluno(a): '))
cidade = input('Cidade: ')

nota = input('Digite as três notas separadas por espaço: ')
lista_notas = [float(n) for n in nota.split()]

cursos = input('Digite os três cursos separados por vírgula: ')
lista_cursos = [c for c in cursos.split(',')]

# Definindo o dicionário
registro = {
    'Nome': nome,
    'Idade': idade,
    'Cursos': lista_cursos,
    'Notas': lista_notas,
    'Cidade': cidade
}

# Calculando a média de notas
media = sum(registro['Notas']) / len(registro['Notas'])

# Imprimindo o relatório do aluno(a)
print('\n--- RELATÓRIO ESCOLAR ---')
print(f'Nome: {nome}')

# Imprimindo todos os cursos
cursos_str = ','.join(registro['Cursos'])
print(f'Cursos: {cursos_str}')
print(f'Média: {media:.2f}')

# Verificando status
situacao = 'Aprovado(a)' if media >= 7 else 'Reprovado(a)' 
print(f'Status: {situacao}')

# Adicionando curso
novo_curso = 'Inglês'
registro['Cursos'].append(novo_curso)

# Imprimindo a lista atualizada
cursos_atualizados = ', '.join(registro['Cursos'])
print(f'Lista de Cursos Atualizada: {cursos_atualizados}')
```

3. **Lógica Condicional**: Aplicação de operadores ternários (`if/else` em uma linha) para determinar o status do aluno de forma concisa.
