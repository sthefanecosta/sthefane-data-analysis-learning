# 📅 QUINTA-FEIRA: Dicionários

**Descrição:** Nesta seção, estudamos a estrutura de dados `dict` em Python, que permite armazenar informações em pares de chave-valor. Aprendemos a criar, acessar, modificar dinamicamente elementos e iterar sobre chaves, valores e itens.

### Exercício 1: Criar e acessar dicionário
Prática de declaração de dicionários e acesso a valores através de chaves específicas.

```python
# Criando o dicionário
meu_computador = {
    'marca': 'Dell',
    'processador': 'Intel Core',
    'memória': '16GB',
    'sistema': 'Windows'
}

# Acessando e exibindo informações
print(meu_computador['marca'])
print(meu_computador['processador'])
print(meu_computador['memória'])
print(meu_computador['sistema'])
```

### Exercício 2: Modificar dicionário
Manipulação de dicionários existentes: adição de novas chaves e atualização de valores.

```python
produto = {
    'nome': 'Café',
    'preço': 15.78,
    'em_estoque': True
}

# Adicionando nova chave
produto['desconto'] = 10

# Alterando o status do produto
produto['em_estoque'] = False

# Imprimindo o dicionário final
print(produto)
```

### Exercício 3: Iterar dicionário
Uso de métodos `.items()`, `.keys()` e `.values()` para percorrer dicionários e processar seus dados.

```python
alunos = {
    'Ana': 10.0,
    'Pedro': 8.5,
    'Luca': 7.6
}

# Percorrendo o dicionário e imprimindo cada aluno com sua nota
print('--- ALUNOS E NOTAS ---')
for k, v in alunos.items():
    print(f'{k}: {v}')

# Imprimindo apenas os nomes
print('\n-- NOMES DOS ALUNOS --')
for nome in alunos.keys():
    print(nome)

# Imprimindo apenas as notas
print('\n-- NOTAS --')

for nota in alunos.values():
    print(nota)
```

### 💡 Notas de Estudo
**O que aprendi de mais importante hoje:**

1. **Estrutura de Dicionários**: Armazenamento de dados baseados em pares `chave: valor`, ideais para representar objetos do mundo real (como produtos ou computadores).
2. **Manipulação Dinâmica**: Facilidade em adicionar (`dict['chave'] = valor`) ou alterar valores existentes durante a execução do programa.
3. **Iteração Eficiente**: Uso dos métodos `.items()` para obter chave e valor simultaneamente, `.keys()` para listar apenas as identificações e `.values()` para processar apenas os dados.
