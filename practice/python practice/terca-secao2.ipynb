### Exercício 1: Converter strings em números
Prática de conversão de tipos (casting) entre texto e números (inteiros e decimais).

```python
# Criando a variável
string = '3.14'
inteiro = 100

# Convertendo a variável
real = float(string)
string_2 = str(inteiro)

# Verificando o tipo de dados
print(f'Variável convertida: {real} | Tipo: {type(real)}')
print(f'Variável convertida: {string_2} | Tipo: {type(string_2)}')
```

### Exercício 2: Calcular preço final com imposto
Aplicação de lógica aritmética composta para precificação e uso de formatação f-string.

```python
# Definindo variáveis
preco_produto = 50.00
quantidade = 3
taxa_imposto = 0.15  # 15%

# Calcular total
subtotal = preco_produto * quantidade
imposto = subtotal * taxa_imposto
total = subtotal + imposto

print(f"Preço unitário: R$ {preco_produto}")
print(f"Quantidade: {quantidade}")
print(f"Subtotal: R$ {subtotal}")
print(f"Imposto (15%): R$ {imposto:.2f}")
print(f"Total: R$ {total:.2f}")
```

### Exercício 3: Calcular média de notas
Cálculo de média aritmética simples com foco na precisão de casas decimais.

```python
# Definindo variáveis
nota1 = 8.0
nota2 = 7.5
nota3 = 9.0

media = (nota1 + nota2 + nota3) / 3
print(f"Notas: {nota1}, {nota2}, {nota3}")
print(f"Média: {media:.2f}")
```
