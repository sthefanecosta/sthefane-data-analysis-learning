### Exercício 1: Tipos de dados básicos
Prática de declaração de variáveis com tipos diferentes e exibição formatada.

```python
# Criando variáveis
nome = 'Sthefane'
idade = 19
cidade = 'Pedro Leopoldo'
altura = 1.73
booleano = True

# Imprimindo frase formatada
print(f'Meu nome é {nome}') 
print(f'Tenho {idade} anos.') 
print(f'Moro em {cidade}.') 
print(f'Tenho {altura} de altura.') 
print(f'Sou estudante? {booleano}')

# Verificando o tipo de cada variável
print(type(nome))
print(type(idade))
print(type(cidade))
print(type(altura))
```

### Exercício 2: Operações com strings
Exploração de métodos de string para transformação e análise de texto.

```python
# Criando a variável que contém a frase
frase = "Python é melhor que Excel"

# Imprimindo a frase com letras maiúsculas
print(frase.upper())

# Verificando quantidades de palavras
print(len(frase.split()))

# Substituindo a palavra 'Excel' e exibindo nova frase
nova_frase = frase.replace('Excel', 'SQL')
print(nova_frase)
```

### Exercício 3: Cálculo de desconto
Aplicação de lógica matemática e formatação de valores monetários.

```python
# Armazenando os dados nas variáveis
preco = 250
desconto = 20

# Calculando o desconto
valor_desconto = (preco * desconto / 100)

# Preço final
preco_final = preco - valor_desconto

# Exibindo mensagem com os dados
print(f'Preço Original: R$ {preco:.2f} | Porcentagem de desconto: {desconto}% | Valor Final: R$ {preco_final:.2f} ')
```
