# **Métodos**

## **📚 Objetivo desta aula**
*Diferenciar funções de métodos e aprender a utilizar as funcionalidades específicas atreladas a tipos de objetos (como listas e strings).*

## **1️⃣ Conceito Teórico**

**O que é?**

*Um método é uma função que pertence a um objeto específico. Enquanto uma função é chamada sozinha (ex: len(lista)), um método é chamado "através" de um objeto, usando um ponto (ex: lista.append()).*

**Por quê aprender?**

Na Ciência de Dados, tudo em Python é um objeto (DataFrames, Séries, Strings). Saber "o que cada objeto pode fazer" (quais métodos ele possui) é o que permite você limpar, filtrar e analisar seus dados.

```python
# Utilizando o metódo index() em fam
fam = ['liz', 1.73, 'emma', 1.68, 'mom', 1.71, 'dad', 1.89]
fam.index('mom')
```

```python
# Utilizando o metódo count
fam.count(1.73)
```

```python
# Metódos para string
sister = 'liz'
print(sister)
```

```python
# Metódos para string
sister.capitalize()
```

```python
# Utilizando replace
sister.replace('z', 'sa')
```

## **Metódos de Listas**

*Use o método .index() para obter o índice do elemento de areas que é igual a 20.0. Imprima esse índice.*

*Chame .count() com areas para descobrir quantas vezes 9.50 aparece na lista. Novamente, basta imprimir esse número.*

```python
# Criando a lista areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Imprimindo o número de elemtnos 20.0 e utilizando index
print(areas.index(20.0))

# Imprimindo quantas areas possui com o metódo count
print(areas.count(9.50))
```

## **PACOTES**

## **📚 Objetivo desta aula**

*Entender como encapsular código em blocos reutilizáveis (funções) e como aproveitar o trabalho de outros programadores através de bibliotecas prontas (pacotes).*

**Pacotes:** *São coleções de funções e scripts criados por terceiros para resolver problemas específicos (como análise de dados ou criação de gráficos).*

**Por quê aprender?**

*Funções mantêm seu código limpo e evitável de erros (DRY - Don't Repeat Yourself). Pacotes permitem que você utilize o poder de ferramentas como pandas ou numpy, que são o padrão ouro na Ciência de Dados.*

*Sempre que estiver instalando algo novo, você usará o **pip install nome_do_pacote** no seu terminal. Já dentro do Jupyter, você usa import para trazer as ferramentas que precisa.*

**Dica:** *Ao importar, você pode dar apelidos para facilitar: **import pandas as pd.** Isso é a convenção universal que você verá em todos os projetos de análise de dados.*

```python
import numpy
array([1, 2, 3])
```

```python
numpy.array([1, 2, 3])
```
