## 📚 Objetivo desta aula
Entender como encapsular código em blocos reutilizáveis (`funções`) e como aproveitar bibliotecas prontas (`pacotes`) para otimizar o trabalho.

## 1️⃣ Funções
* **O que são?** Blocos de código que realizam tarefas específicas. Em vez de repetir código, você `chama` a função.
* **Por que usar?** Segue o princípio `DRY` (Don't Repeat Yourself), mantendo seu código limpo e com menos chances de erro.
* **Exemplos:** `max()`, `round()`, `type()`, `len()`, `sorted()`. Use `help(nome_da_função)` para documentação.

## 2️⃣ Métodos
* **O que são?** Funções que pertencem a um objeto específico. Enquanto funções são chamadas isoladamente (ex: `len(lista)`), métodos são chamados através de um objeto usando um ponto (ex: `lista.append()`).
* **Importância:** Em Ciência de Dados, tudo é um objeto. Saber quais métodos cada objeto possui permite que você manipule dados com eficiência.

## 3️⃣ Pacotes
* **O que são?** Coleções de funções e scripts criados por terceiros.
* **Uso:** Utilize `pip install nome_do_pacote` no terminal para instalar e `import` no Jupyter para usar.
* **Dica:** Use apelidos para facilitar o uso (ex: `import numpy as np`).

```python
fam = [1.73, 1.68, 1.71, 1.89]

# Exemplos de uso de funções
print("Máximo:", max(fam))
print("Arredondamento:", round(1.68, 1))

# Verificando ajuda da função
help(round)
```

```python
first = [11.25, 18.0, 20.0]
second = [10.75, 9.50]

full = first + second
full_sorted = sorted(full, reverse=True)
print(full_sorted)
```

```python
fam = ['liz', 1.73, 'emma', 1.68, 'mom', 1.71, 'dad', 1.89]
sister = 'liz'

print(fam.index('mom'))
print(fam.count(1.73))
print(sister.capitalize())
print(sister.replace('z', 'sa'))
```
