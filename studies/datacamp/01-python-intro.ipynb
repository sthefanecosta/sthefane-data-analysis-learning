## 1️⃣ Conceito Teórico

### O que são Scripts?
Scripts Python são arquivos de texto com a extensão `.py`. Eles funcionam como uma lista organizada de comandos que o interpretador Python executa sequencialmente.

### Por que aprender esta abordagem?
* **Memorização:** Escrever scripts ajuda a fixar a sintaxe.
* **Organização:** Permite salvar seu trabalho para uso posterior.
* **Interatividade:** O shell permite testar partes do código antes de consolidá-las no script final.

---

## 2️⃣ Exemplos Práticos de Operações
O Python funciona como uma calculadora potente. Abaixo, os principais operadores aritméticos:

* **Divisão:** `5 / 8` = `0.625`
* **Adição:** `4 + 5` = `9`
* **Subtração:** `10 - 5` = `5`
* **Multiplicação:** `3 * 5` = `15`

> **Nota sobre o `print()`:** O comando `print()` é essencial para exibir a saída dos cálculos no console. Sem ele, em scripts salvos, o resultado do cálculo não aparece visualmente para o usuário.

---

### 💡 Resumo do Aprendizado
O Python processa as operações matemáticas seguindo a lógica de comandos. A utilização do `print()` garante que o fluxo de dados seja visível, transformando cálculos abstratos em informações legíveis na tela.

---

```python
# Exemplos de testes aritméticos utilizados na aula:

print("Divisão (5 / 8):", 5 / 8)
print("Adição (4 + 5):", 4 + 5)
print("Divisão (10 / 2):", 10 / 2)
print("Subtração (10 - 5):", 10 - 5)
print("Multiplicação (3 * 5):", 3 * 5)
```
