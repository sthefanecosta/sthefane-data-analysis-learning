## 📚 Objetivo desta aula
Aprender a criar listas `[]`, agrupar diferentes tipos de dados em uma variável e realizar operações de acesso e manipulação.

## 1️⃣ Conceito Teórico
* **O que é?** Estrutura ordenada e mutável que armazena sequências de elementos (podendo ser tipos mistos).
* **Por que aprender?** Essencial para organizar conjuntos de dados, como listas de preços ou observações de experimentos.

> **Dica de Ouro:** A contagem começa no `0`. O primeiro elemento está no índice `0`. Índices negativos acessam a lista de trás para frente (o `-1` é o último).

## 2️⃣ Exemplos da Aula
* **Lista Simples:** `fam = ['Liz', 1.73, 'Emma', 1.68]`
* **Lista de Listas:** `fam2 = [['Liz', 1.73], ['Emma', 1.68]]`
* **Slicing:** `lista[início : fim]` (o `fim` é exclusivo).

## 3️⃣ Manipulação de Listas
* **Alterar valor:** `fam[0] = 'Lisa'`
* **Deletar:** `del fam[2]`
* **Cópia Segura:** `y = x[:]` ou `y = list(x)` para evitar que alterações em `y` afetem `x`.

```python
# Listas em Python
fam = ['Liz', 1.73, 'Emma', 1.68, 'mom', 1.71, 'dad', 1.89]
print(fam)

# Lista de listas
fam2 = [['Liz', 1.73], ['Emma', 1.68], ['mom', 1.71], ['dad', 1.89]]
print(fam2)
```

```python
hall = 11.25; kit = 18.0; liv = 20.0; bed = 10.75; bath = 9.50

# Lista areas
areas = [hall, kit, liv, bed, bath]
print(areas)

# Lista com strings e valores
areas_info = ['hallway', hall, 'kitchen', kit, 'living room', liv, 'bedroom', bed, 'bathroom', bath]
print(areas_info)
```

```python
fam = ['Liz', 1.73, 'Emma', 1.68, 'mom', 1.71, 'dad', 1.89]

print(fam[3])    # Índice 3
print(fam[-1])   # Último item
print(fam[3:5])  # Fatiamento
```

```python
# Alterando e deletando
fam[7] = 1.86
del fam[2]

# Cópia segura
x = ['a', 'b', 'c']
y = x[:] 
y[1] = 'z'
print(f"X original: {x}, Y copiado: {y}")
```
