## 1️⃣ Docstrings
* **O que é?** Uma string colocada logo após a definição da função (`def`).
* **Sintaxe:** Usa-se aspas triplas (`"""`).
* **Diferencial:** Diferente de comentários (`#`), a `docstring` é lida pelo Python e pode ser consultada via `help()`.
* **Estrutura recomendada:**
    * **Resumo:** O que a função faz.
    * **Args:** Parâmetros de entrada e seus tipos.
    * **Returns:** O que a função entrega.
    * **Raises:** Erros intencionais levantados.

## 2️⃣ Conceitos Essenciais
* **DRY (Don't Repeat Yourself):** Se você está escrevendo o mesmo código várias vezes, transforme-o em uma função.
* **Do One Thing:** Uma função deve ter um objetivo único. Se ela calcula *e* imprime, divida em duas funções.
* **Mutabilidade:** * **Listas (Mutáveis):** Alterar uma lista dentro de uma função altera o objeto original. Use `lista[:]` para criar uma cópia segura.
    * **Inteiros/Strings (Imutáveis):** Não alteram o original; a função apenas cria uma nova referência.

````python
# Bar (Escopo)
def bar(x):
    x = x + 90

my_var = 3
bar(my_var)
my_var
````

```python 
    # Calcular preço final para 3 clientes

## 📋 EXERCÍCIO 3: Dividir uma Função (Do One Thing)
*Você tem uma função que calcula a média e imprime o resultado:*
1. Divida em 2 funções: Uma que calcula a média (retorna valor) e outra que verifica aprovação (retorna True/False).
2. Teste as funções separadamente e depois combine para processar um aluno.

````python
# Criando as funções
def calcular_media(*notas):
    """Calcula a média de uma quantidade váriavel de números.
    Args:
        *notas(float ou int): Números para calcular a média
    Returns:
        media(float): Média calculada.
    Raises:
        ValueError: Se a lista fornecida estiver vazio, não é possível dividir por zero.
    """
    if len(notas) == 0:
        return 0
        
    media = sum(notas) / len(notas)
    return media

def situacao(nota, media_corte=7.0):
    """Define a situação do aluno como Aprovado(a) ou Reprovado(a).
    Args:
        nota(float ou int): Nota obtida pelo aluno
        media_corte(float ou int, opcional): Nota miníma definida para aprovação. Default=7.0
    Returns:
        'Aprovado(a)' se a nota for maior ou igual ao corte, 'Reprovado(a)' se a nota for menor ao corte.
    Raises:
        ValueError: Se algum dos parâmetros fornecidos forem negativos ou maior que 10. 
    """
    if not (0 <= nota <= 10) or not (0 <= media_corte <= 10):
        raise ValueError('Nota e média de corte deve estar entre 0 e 10.')
    if nota >= 7.0:
        return 'Aprovado(a)'
    else:
        return 'Reprovado(a)'

# Testando a função
media = calcular_media(4.5, 5.7, 8.0)
status = situacao(media)

print(f'Média: {media:.2f} | Status: {status}')
````

## 📋 EXERCÍCIO 4: Argumentos Padrão
1. Complete a função para calcular preço com desconto (padrão 10%).
2. Teste COM o desconto padrão.
3. Teste COM desconto diferente.

````python
def calcular_preco_desconto(preco, desconto=0.10):
    """ Calcula o preço com desconto.
    Args:
        preco(float): Preço fornecido.
        desconto(float, opcional): Desconto fornecido. Default=0.10
    Returns:
        preco_desconto(float): Valor do preço calculado com o desconto.
    Raises:
        ValueError: Se o valor fornecido não for um número.
    """
    valor_desconto = preco * desconto
    preco_desconto = preco - valor_desconto
    return preco_desconto

teste_1 = calcular_preco_desconto(250, 0.5)
teste_2 = calcular_preco_desconto(90)
teste_3 = calcular_preco_desconto(150, 0.40)

print(f'Teste 1 | Valor: R$ {teste_1:.2f}')
print(f'Teste 2 | Valor: R$ {teste_2:.2f}')
print(f'Teste 3 | Valor: R$ {teste_3:.2f}')
````

````python
# Função reescrita
def adicionar_produto(lista_compras, produto):
    novas_compras = lista_compras[:]
    novas_compras.append(produto)
    return novas_compras

minhas_compras = ["pão", "leite"]
resultado = adicionar_produto(minhas_compras, "queijo")

print("minhas_compras:", minhas_compras)
print("resultado:", resultado)
````

## 📋 EXERCÍCIO 7: Projeto Integrador
*Sistema de cálculo de salário:*
1. Crie a função que recebe nome, salário e alíquota (padrão 15%).
2. Use docstring, valide se salário > 0 e retorne um dicionário com os dados.

````python
# Definindo a função
def calcular_salario_imposto(nome, salario, aliquota=0.15):
    """ Calcula o valor do salário com o imposto e retorna um dicionário.
    Args:
        nome(str): Nome do cliente fornecido.
        salario(float): Salário Bruto fornecido.
        aliquota(float, opcional): Valor do imposto fornecido. Default = 0.15
    Returns:
        Dict: Dicionário com os dados: nome, salário bruto, valor do imposto e valor líquido.
    Raises:
        ValueError: Se um dos valores fornecidos forem menor que zero.
    """
    if salario <= 0:
        raise ValueError('Salário Bruto deve ser maior que zero.')

    valor_imposto = salario * aliquota
    valor_salario = salario - valor_imposto

    return {
        'nome': nome,
        'salário bruto': salario,
        'imposto': valor_imposto,
        'líquido': valor_salario
    }

print(calcular_salario_imposto("Sthefane", 2000)) # Com imposto padrão
print(calcular_salario_imposto("Ana", 3000, 0.20)) # Com imposto customizado
````
