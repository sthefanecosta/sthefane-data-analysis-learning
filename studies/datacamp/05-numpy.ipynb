## **1️⃣ Conceito Teórico**
**O que é?**

*O NumPy é uma biblioteca que introduz o objeto ndarray (n-dimensional array). Ao contrário de uma lista, que pode conter vários tipos de dados, o array do NumPy é otimizado para conter elementos de um único tipo (geralmente números), permitindo operações matemáticas em toda a estrutura de uma vez.*

**Por quê aprender?**

*Listas comuns do Python são lentas para cálculos grandes. O NumPy utiliza C "por baixo dos panos", o que torna as operações matemáticas milhares de vezes mais rápidas. É a base para o Pandas, Matplotlib e quase toda a Inteligência Artificial.*

```python
import numpy as np
```

```python
# Criando as listas
altura = [1.73, 1.68, 1.71, 1.89, 1.79]
peso = [65.4, 59.2, 63.6, 88.4, 68.7]
```

```python
# Tranformando as listas em arrays
np_altura = np.array(altura)
np_altura
```

```python
np_peso = np.array(peso)
np_peso
```

```python
# Fazendo o cáalculo de IMC
imc = np_peso / np_altura ** 2
imc
```
