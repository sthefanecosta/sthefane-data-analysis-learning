## 1️⃣ Conceito Teórico
* **O que é o NumPy?** É uma biblioteca poderosa que introduz o objeto `ndarray` (n-dimensional array).
* **Arrays vs. Listas:** * Listas podem conter tipos de dados variados e são flexíveis, mas lentas para cálculos.
    * Arrays do NumPy são otimizados para conter elementos de um **único tipo** (homogêneos), permitindo operações matemáticas vetorizadas (aplicadas a todos os elementos de uma vez).
* **Performance:** O NumPy processa cálculos em C "por baixo dos panos", tornando operações matemáticas milhares de vezes mais rápidas que em listas padrão. É o alicerce para ferramentas como `Pandas` e `Scikit-Learn`.

> **Atenção:** Ao somar listas, você as concatena (`[1,2] + [1,2]` vira `[1,2,1,2]`). Ao somar arrays, você realiza a operação aritmética elemento a elemento (`np.array([1,2]) + np.array([1,2])` vira `[2,4]`).

```python
import numpy as np

# Criando as listas base
altura = [1.73, 1.68, 1.71, 1.89, 1.79]
peso = [65.4, 59.2, 63.6, 88.4, 68.7]
```

```python
# Transformando listas em arrays do NumPy
np_altura = np.array(altura)
np_peso = np.array(peso)

print(f"Altura (array): {np_altura}")
print(f"Peso (array): {np_peso}")
```

```python
# Cálculo do IMC utilizando vetorização (operação elemento a elemento)
imc = np_peso / np_altura ** 2
print(f"IMC: {imc}")
```

```python
# NumPy força um tipo único (neste caso, string)
tipagem_unica = np.array([1.0, 'is', True])
print(f"Array homogêneo: {tipagem_unica}")

# Diferença de comportamento na soma
python_list = [1, 2, 3]
numpy_array = np.array([1, 2, 3])

print("Soma de listas (concatenação):", python_list + python_list)
print("Soma de arrays (aritmética):", numpy_array + numpy_array)
```
