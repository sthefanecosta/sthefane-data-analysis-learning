## **1️⃣ Conceito Teórico**

**O que é?**

Uma lista é uma estrutura que permite armazenar uma sequência de elementos (podendo ser de tipos diferentes) usando colchetes []. É uma coleção ordenada e mutável.

**Por quê aprender?**

Na análise de dados, quase nunca trabalhamos com um valor único. Listas permitem organizar conjuntos de observações, como uma lista de preços, nomes de produtos ou resultados de experimentos.

## **2️⃣ Exemplos da Aula**

```python
# Listas em Python
fam = ['Liz', 1.73, 'Emma', 1.68, 'mom', 1.71, 'dad', 1.89]
fam
```

```python
# Lista de listas

fam2 = fam = [['Liz', 1.73], 
              ['Emma', 1.68], 
              ['mom', 1.71], 
              ['dad', 1.89]]
fam2
```

**Dica importante:** 

Lembre-se sempre: em Python, a contagem começa no zero. O primeiro elemento da lista está no índice 0, o segundo no índice 1, e assim por diante. Se tentar acessar um índice que não existe, o Python emitirá um erro (IndexError).

# **Subdivisão de Listas**

## **📚 Objetivo desta aula**

**Aprender a acessar informações de uma lista utilizando índice, sem precisar criar uma nova variável manualmente.**

## **1️⃣ Conceito Teórico**

**O que é?** 

O slicing permite acessar um intervalo de itens em uma lista usando a sintaxe lista[início : fim]. O número que você coloca no "início" é incluído, mas o número no "fim" é excluído.

**Por quê aprender?**

Crucial para manipulação de dados. Quando você tiver uma lista com 1.000 registros, poderá facilmente extrair apenas os primeiros 10, os últimos 5 ou qualquer intervalo específico necessário para sua análise.


## **2️⃣ Exemplos da Aula**

```python
fam = ['Liz', 1.73, 'Emma', 1.68, 'mom', 1.71, 'dad', 1.89]
fam
```

## **[início : fim]**

```python
fam[3]
```

```python
fam[6]
```

```python
fam[-1]
```

```python
fam[3:5]
```

## **Exercícios:**

Imprima o segundo elemento da lista areas (ele tem o valor 11.25).

Crie um subconjunto e imprima o último elemento de areas, que é 9.50. Aqui faz sentido usar um índice negativo!

Selecione o número que representa a área da sala de estar (20.0) e imprima-o.


```python
areas = ["hallway", 11.25, "kitchen", 18.0, "living room", 20.0, "bedroom", 10.75, "bathroom", 9.50]

print(areas[1])

print(areas[-1])

print(areas[5])
```

Use o fatiamento para criar uma lista, downstairs, que contém os primeiros 6 elementos de areas.

Crie upstairs, com os últimos elementos de 4 de areas. Desta vez, simplifique o fatiamento omitindo o índice final (end).

Imprima downstairs e upstairs usando print().


```python
downstairs = areas[:6]

upstairs = areas[-4:]

print(downstairs)
print(upstairs)
```

# **Manipulação de listas**

Aprender a modificar, remover e alterar elementos dentro de uma lista existente para manter seus dados atualizados.

## **1️⃣ Conceito Teórico**

**O que é?**

Manipulação de listas envolve o uso de métodos (funções internas) para mudar o conteúdo de uma lista após ela já ter sido criada. Podemos substituir valores por índice, deletar itens ou trocar elementos de lugar.

**Por quê aprender?**

Em projetos reais, os dados chegam "sujos" ou incompletos. Você precisará constantemente atualizar valores errados, remover duplicatas ou inserir novas informações em listas que já existem.

```python
# Lista Inicial
fam = ['Liz', 1.73, 'Emma', 1.68, 'mom', 1.71, 'dad', 1.89]
fam
```

```python
# 1. Alterar um valor específico
fam[7] = 1.86
fam
```

```python
# Alterando uma lista de listas
fam[0:2] = ['Lisa', 1.74]
fam
```

```python
# Incluindo elementos 
fam + ['me', 1.73]
```

```python
# Excluindo elementos
del fam[2]
fam
```

```python
x = ['a', 'b', 'c']

y = x
y[1] = 'z'
y
```

```python
x
```

```python
# Copiando listas:

# Opção 1:
y = list(x)

# Opção 2:
y = x[:]

y[1] = 'z'
x
```

Altere o segundo comando, que cria a variável areas_copy, de modo que areas_copy seja uma cópia explícita de areas.

```python
# Criando a lista areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Alterando comando
areas_copy = areas[:]

# Alterando areas_copy
areas_copy[0] = 5.0

# Imprimindo areas
print(areas)
print(areas_copy)
```
