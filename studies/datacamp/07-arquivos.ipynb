## 1. O Fluxo de Trabalho (A Regra do `with`)
Sempre que abrir um arquivo, use o gerenciador de contexto `with`. Ele fecha o arquivo automaticamente, garantindo que os dados sejam gravados no disco e evitando corrupção.

* **Modos de abertura:**
    * `'r'` (Read): Leitura (padrão).
    * `'w'` (Write): Escrita (cria um arquivo ou sobrescreve se já existir).
    * `'a'` (Append): Adição (escreve no final do arquivo sem apagar o que já existe).
* **Dica:** Sempre utilize `encoding='utf-8'` para suportar acentuação e caracteres especiais corretamente.

---

## 2. Exemplo Prático de Escrita e Leitura
```python
# Escrita segura
with open("estudo.txt", "w", encoding="utf-8") as f:
    f.write("Linha 1: Python é eficiente!\n")
    f.write("Linha 2: O 'with' evita arquivos corrompidos.\n")

# Leitura segura
with open("estudo.txt", "r", encoding="utf-8") as f:
    print(f.read())

## 📓 Exercício 1: Registro de Contatos
Crie uma função `salvar_contato(nome, telefone)` que adicione um novo registro no arquivo `agenda.txt`. 

**Regras:**
1. O arquivo deve ser aberto em modo de adição (`'a'`).
2. O nome e telefone devem ser escritos no formato `Nome: Telefone` (seguido de quebra de linha).
3. Garanta que o arquivo seja fechado corretamente usando o gerenciador de contexto `with`.

````python
def salvar_contato(nome, telefone):
    """Adiciona um contato ao arquivo 'agenda.txt'.

    Args:
        nome(str): Uma string com o nome do contato a ser salvo.
        telefone(str): Telefone associado ao nome do contato.
        
    Side Effects:
        Escreve uma nova linha no arquivo 'agenda.txt'
    """
    with open("agenda.txt", "a", encoding="utf-8") as f:
        f.write(f"{nome} : {telefone}\n")

# Chamando a função para adicionar contatos
salvar_contato("Ana", "9999-8888")
salvar_contato("Luis", "8888-7777")
salvar_contato("Carol", "7777-6666")

print("Contatos adicionados com sucesso!")
````

## 📓 Exercício 2: Leitura de Log
Crie uma função `ler_log(nome_arquivo)` que abre um arquivo de texto e imprime na tela apenas as linhas que começam com a palavra "ERRO".

**Dica:** - Você pode ler o arquivo linha por linha usando um loop `for linha in f:`.
- Use o método `linha.startswith("ERRO")` para verificar o conteúdo.

````python
def ler_log(nome_arquivo):
    """Veifica e imprime na tela apenas as linhas que começam com a palavra 'ERRO'.

    Args:
        nome_arquivo(str): Arquivo que deve ser aberto.
    Side Effects:
        Imprime as linhas filtradas diretamente no console.
    """
    with open(nome_arquivo, "r", encoding="utf-8") as f:
        for linha in f:
            if linha.startswith("ERRO"):
                print(linha.strip())
                
# Chamada de teste
print("\n--- Logs de Erro ---")
ler_log("sistema.log")
````

## 📓 Exercício 3: Contador de Linhas
Crie uma função `contar_linhas(nome_arquivo)` que abre um arquivo de texto e retorna o número total de linhas existentes nele.

**Requisito:**
- Certifique-se de que a função funcione mesmo se o arquivo for muito grande, processando-o de forma eficiente.

````python
def contar_linhas(nome_arquivo):
    """Conta a quantidade de linhas de um arquivo.

    Args:
        nome_arquivo(str): O arquivo que deve ser aberto.
        
    Returns:
    total_linhas(int): Quantidade de linhas que contém no arquivo.   
    """
    total_linhas = 0
    
    with open(nome_arquivo, "r", encoding="utf-8") as f:
        for linha in f:
            total_linhas += 1
            
    return total_linhas

# Chamada de teste
total = contar_linhas("agenda.txt")
print(f"\nO total de linhas no arquivo é: {total}")
````

## 📓 Exercício 4: Copiadora de Arquivos
Crie uma função `copiar_arquivo(origem, destino)` que leia todo o conteúdo de um arquivo de origem e crie um novo arquivo de destino contendo exatamente a mesma informação.

**Dica:** - Abra o arquivo de origem em modo `'r'` (leitura).
- Abra o arquivo de destino em modo `'w'` (escrita).

````python
def copiar_arquivo(origem, destino):
    """Faz a leitura de um arquivo e armazena uma cópia no arquivo destino.

    Args:
        origem(str): Arquivo de origem.
        destino(str): Arquivo de destino para a cópia
    Side Effects:
        Gera uma cópia do arquivo origem em um arquivo destino.
    """
    with open(origem, "r", encoding="utf-8") as f_origem, open(destino, "w", encoding="utf-8") as f_destino:
        for linha in f_origem:
            f_destino.write(linha)
            
# Chamada de teste
copiar_arquivo("agenda.txt", "agenda_backup.txt")

# Verificação extra
print(f"Cópia concluída. O backup tem {contar_linhas('agenda_backup.txt')} linhas.")
````
