---

## 1️⃣ Conceito Teórico

### O que são Variáveis?
Variáveis são como "caixas" com nomes que usamos para guardar valores. Os tipos de dados definem a natureza desses valores (números inteiros, decimais ou texto).

### Por que aprender?
Na análise de dados, você manipulará diversas informações (datas, valores monetários, nomes). Identificar o tipo correto de dado evita erros cruciais em cálculos e visualizações gráficas.

> **Dica de Ouro:** Use a função `type()` sempre que tiver dúvida sobre como o Python está interpretando um dado. Ela confirmará se o conteúdo é `int`, `float`, `str` ou `bool`.

---

## 2️⃣ Exemplos de Tipos de Dados

* **Float (`float`):** Números decimais.
    * *Exemplo:* `altura = 1.79`
* **String (`str`):** Sequência de caracteres (texto).
    * *Exemplo:* `x = 'Isso é uma string'`
* **Booleano (`bool`):** Valores lógicos de Verdadeiro ou Falso.
    * *Exemplo:* `z = True`

---

## 3️⃣ Comportamento dos Tipos
O tipo do dado muda a forma como o Python processa a informação:
* **Soma numérica:** `2 + 3` resulta em `5` (cálculo matemático).
* **Concatenação de strings:** `'ab' + 'cd'` resulta em `'abcd'` (união de textos).

---

```python
# 1. Variáveis e Cálculo de IMC
altura = 1.79
peso = 68.7
imc = peso / altura ** 2
print(f"IMC: {imc}")

# 2. Verificação de tipos
print("Tipo da altura:", type(altura))
print("Tipo de uma string:", type('Isso é uma string'))
print("Tipo de um booleano:", type(True))

# 3. Diferença de comportamento entre tipos
print("Soma de números (2+3):", 2 + 3)
print("Concatenação de strings ('ab'+'cd'):", 'ab' + 'cd')

# 4. Exercício de Poupança (Savings)
savings = 100
monthly_savings = 10
num_months = 4
new_savings = monthly_savings * num_months
total_savings = savings + new_savings

print(f"Total na poupança: {total_savings}")
print("Tipo do total:", type(total_savings))
```
