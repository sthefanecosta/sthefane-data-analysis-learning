# 📚 Objetivo desta aula

Aprender a nomear e atribuir valores a variáveis e identificar os diferentes tipos de dados básicos no Python.


# 1️⃣ Conceito Teórico

**O que é?**

Variáveis são "caixas" com nomes que usamos para guardar valores. Tipos de dados definem a natureza desses valores (números inteiros, números decimais ou texto).

**Por quê aprender?**

Na análise de dados, você precisa manipular diferentes tipos de informações (como datas, valores monetários ou nomes de clientes). Saber o tipo de dado correto evita erros nos seus cálculos e gráficos.
